# Explore Metadata to isolate specific cell populations

for more informative or less confounded conclusions

```python
# Load Meta Data
import os
import pandas as pd

save_path = '/media/rohola/ssd_storage/primary_metadata/'; print(len(os.listdir(save_path)))
files = sorted(os.listdir(save_path))
df = pd.concat([pd.read_pickle(save_path+f) for f in files], ignore_index=True)
```

```python
import json
project_path = "/media/lleger/LaCie/mit/disease_geometry/"
diseases = json.load(open(project_path + 'diseases.json'))['diseases']
print(diseases)
diseases = ['small cell lung carcinoma', 'lung large cell carcinoma', 'non-small cell lung carcinoma', 'glioblastoma', 'lung adenocarcinoma',
             ] + ['Parkinson disease', 'Alzheimer disease', 'dementia', 'Lewy body dementia'] + ["dilated cardiomyopathy", "myocardial infarction"] + ['chronic kidney disease', "clear cell renal carcinoma"]
            
dtest = df[int(0.7*len(df)):]
for disease in diseases:
    print(disease)
    display(dtest[dtest['disease'] == disease][['cell_type']].value_counts().reset_index().head(10))
```
